```python
# Import the tools we need
from sklearn.datasets import load_iris  # Gets the iris flower data
from sklearn.model_selection import train_test_split  # Splits data into practice and test sets
from sklearn.linear_model import LogisticRegression  # The "brain" that learns patterns
import joblib  # Saves our trained brain to a file

# Get the iris flower dataset (measurements and types)
iris = load_iris()

# Split the data: 80% for teaching the brain, 20% for testing it
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

# Create a new "brain" that will learn to recognize flowers
model = LogisticRegression(max_iter=200)

# Teach the brain by showing it flower measurements and their correct types
model.fit(X_train, y_train)

# Save the trained brain to a file so we can use it later without retraining
joblib.dump(model, r'C:\Users\user\Desktop\Python Course\AI-ML-Python-Scripts\AI & ML Models\iris_model.pkl')
```

```python
# Import tools to build a web service
from flask import Flask, request, jsonify  # Flask creates websites, request reads incoming data, jsonify formats responses
import joblib  # Loads our saved brain
import numpy as np  # Handles number arrays

# Create a web application
app = Flask(__name__)

# Load the trained brain from the file we saved earlier
model = joblib.load('iris_model.pkl')

# Create a webpage at "/predict" that accepts questions
@app.route('/predict', methods=['POST'])
def predict():
    # Get the flower measurements someone sent us
    data = request.get_json(force=True)
    
    # Ask the brain what type of flower this is
    prediction = model.predict(np.array(data['input']).reshape(1, -1))
    
    # Send back the answer (0, 1, or 2 for different flower types)
    return jsonify({'prediction': int(prediction[0])})

# Start the web service so anyone can connect and ask questions
if __name__ == '__main__':
    app.run(host='0.0.0.0', port=80)
```
